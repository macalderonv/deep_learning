```python
eje_x = []
eje_y = []
eje_z = []

with open('Acc_Data_2018_10_19_12_04_19_P.txt') as archivo:
    
    for linea in archivo:
        #print(linea)
        if linea[1] == 't':
            print('Tiempo:' + linea[3:15])
            print('Dispositivo: ' + linea[34:38])
            eje_x.append(linea[linea.find("X")+2:linea.find("Y")-3])
            eje_y.append(linea[linea.find("Y")+2:linea.find("Z")-3])
            eje_z.append(linea[linea.find("Z")+2:linea.find("q")-2])
            #print('Eje X: ' + linea[linea.find("X")+2:linea.find("Y")-3])
            #print('Eje Y: ' + linea[linea.find("Y")+2:linea.find("Z")-3])
            #print('Eje Z: ' + linea[linea.find("Z")+2:linea.find("q")-2])
        print('---')
        
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

print(len(eje_x))
print(len(eje_y))
fs = 50 #Hz
t = np.arange(0,len(eje_x))/fs
plt.figure(figsize=(8,10))
plt.plot(t,eje_x)
plt.plot(t,eje_y)
plt.plot(t,eje_z)
```
